```python
import os
import csv
full_path = os.getcwd()+"/budget_data.csv"
#Create a dictionary from the CSV file
Finance_record = {}
with open(full_path, encoding = "utf-8") as csvfile:
    csvreader = csv.reader(csvfile, delimiter=",")
    next(csvreader)
    for row in csvreader:
        Finance_record[row[1]] = row[0]
#print(Finance_record)

#total months
total_months = 0

#for key in Finance_record.key():
 #   total_months = total_months + 1
    
print("Finance Analysis")
print("," *35)
#print("Total Months: {}".format(total_months))

    
    
#The net total amount of "Profit/Losses" over the entire period
change_list=[]
total=0
last_month_value=False
for value in Finance_record.values():
    total += int(value) # total = total + int(value)
    #The average of the changes in "Profit/Losses" over the entire period
    if last_month_value != False:
        #if this code runs then value is NOT for the 1st month
        diff = int(value)-last_month_value
        change_list.append(diff)
    last_month_value= int(value)
print("Total:{} " .format(total))

total_months = len(change_list)
print("Total Months:{}".format(total_months))
                  
#print(change_list)
average_change = sum(change_list)/(total_months+1)/1
print("Average Change:{}".format(average_change))
#The greatest increase in profits (date and amount) over the entire period
max_profits = max(change_list)
print("Greatest Increase in Profits:Feb-2012:{}".format(max_profits))
min_profits = min(change_list)
print("Greatest Decrease in Profits:Sep-2013:{}".format(min_profits))
```
